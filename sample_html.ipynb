```python
from html.parser import HTMLParser

class TopicParser(HTMLParser):
    def __init__(self, topic_title):
        super().__init__()
        self.topic_title = topic_title.lower()
        self.topic_information = []
        self.in_topic = False

    def handle_starttag(self, tag, attrs):
        # Check if the tag is a heading tag and if its text contains the topic title
        if tag in ['h1', 'h2', 'h3']:
            for attr, value in attrs:
                if attr == 'id' and self.topic_title in value.lower():
                    self.in_topic = True
                    break

    def handle_endtag(self, tag):
        if tag in ['h1', 'h2', 'h3'] and self.in_topic:
            self.in_topic = False

    def handle_data(self, data):
        if self.in_topic:
            self.topic_information.append(data.strip())

def extract_topic_information(html_file_path, topic_title):
    with open(html_file_path, 'r', encoding='ISO-8859-1') as file:
        html_content = file.read()

    parser = TopicParser(topic_title)
    parser.feed(html_content)
    return parser.topic_information

# Example usage:
html_file_path = 'C360_sample.html'  # Replace 'your_html_file.html' with the path to your HTML file
topic_title = 'Purpose and Objectives'

topic_information = extract_topic_information(html_file_path, topic_title)

if topic_information:
    for info in topic_information:
        print(info)
else:
    print("Topic not found or no information available.")
```

```python
from bs4 import BeautifulSoup

def extract_topic_info_from_html_file(file_path, topic):
    # Read the HTML content from the file
    with open(file_path, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the span element containing the topic
    topic_element = soup.find('span', string=topic)

    # If the topic element is found, find all following sibling elements and extract their text
    if topic_element:
        print("Topic element found:", topic_element)
        # Get all following sibling elements until the next header element is encountered
        topic_info = ''
        next_elements = topic_element.find_next_siblings()
        for element in next_elements:
            if element.name == 'h2':  # Stop when the next topic is encountered
                break
            if element.name == 'p' and element.string:  # Consider only <p> tags with text content
                topic_info += str(element.get_text(strip=True)) + ' '

        return topic_info.strip()
    else:
        return "Topic not found"

# Example usage
file_path = "test1.html"
topic = "Target Audience"
topic_info = extract_topic_info_from_html_file(file_path, topic)
print("Topic Info:", topic_info)
```

```python
from bs4 import BeautifulSoup

def extract_info_between_topics(file_path, start_topic, end_topic):
    with open(file_path, 'r', encoding='utf-8') as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, 'html.parser')

    start_topic_element = soup.find('span', text=start_topic)
    end_topic_element = soup.find('span', text=end_topic)

    content_between = ""

    if start_topic_element and end_topic_element:
        all_elements = soup.find_all(True)

        start_index = all_elements.index(start_topic_element)
        end_index = all_elements.index(end_topic_element)

        for elem in all_elements[start_index + 1:end_index]:
            if elem.name in ['p', 'ul', 'ol']:  
                content_between += elem.get_text(strip=True) + '\n\n'

    return content_between.strip()

# Example usage
file_path = "test1.html"
start_topic = "Key Features and Capabilities"
end_topic = "Expected Outcomes and Benefits"
info = extract_info_between_topics(file_path, start_topic, end_topic)
print(info)
```
